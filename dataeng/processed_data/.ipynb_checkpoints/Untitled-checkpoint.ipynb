```python
import os
path_to_folder = "../02-src-data/"
path_to_img = "https://raw.githubusercontent.com/provectus/internship/main/dataeng/02-src-data/"
src_data = os.listdir(path_to_folder)
```

```python
data_csv, data_img, user_ids = set(), set(), set()
for file in src_data:
    if file.endswith(".csv"):
        user_ids.add(file[:-4])
        data_csv.add(file)
    else:
        data_img.add(file)
```

```python
import pandas as pd
output_data = pd.DataFrame(columns=['id'])
for csv in data_csv:
    output_data.append({'id': csv[:-4]}, ignore_index=True)
    
    csv_file = pd.read_csv(path_to_folder + csv)
#     row = pd.DataFrame(csv_file)
    output_data = output_data.append(csv_file, ignore_index=True)
```
